```python
import cv2
import numpy as np
import os
import time
import threading
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torchvision.transforms as transforms
from torchvision import models

# Caminho para os arquivos
labels_path = r'C:\Users\ulyss\OneDrive - AllTech\Área de Trabalho\Eyes\yolo\coco.names'

# Verificar se o arquivo de rótulos existe
if not os.path.exists(labels_path):
    print(f"Arquivo de rótulos não encontrado: {labels_path}")
    exit()

# Carregar os nomes das classes
with open(labels_path, 'r') as f:
    classes = f.read().strip().split('\n')

# Dicionário de tradução para português
translation_dict = {
    "person": "pessoa",
    "bicycle": "bicicleta",
    "car": "carro",
    "motorbike": "moto",
    "aeroplane": "avião",
    "bus": "ônibus",
    "train": "trem",
    "truck": "caminhão",
    "boat": "barco",
    "traffic light": "semáforo",
    "fire hydrant": "hidrante",
    "stop sign": "placa de pare",
    "parking meter": "parquímetro",
    "bench": "banco",
    "bird": "pássaro",
    "cat": "gato",
    "dog": "cachorro",
    "horse": "cavalo",
    "sheep": "ovelha",
    "cow": "vaca",
    "elephant": "elefante",
    "bear": "urso",
    "zebra": "zebra",
    "giraffe": "girafa",
    "backpack": "mochila",
    "umbrella": "guarda-chuva",
    "handbag": "bolsa",
    "tie": "gravata",
    "suitcase": "mala",
    "frisbee": "frisbee",
    "skis": "esquis",
    "snowboard": "snowboard",
    "sports ball": "bola de esportes",
    "kite": "pipa",
    "baseball bat": "taco de beisebol",
    "baseball glove": "luva de beisebol",
    "skateboard": "skate",
    "surfboard": "prancha de surf",
    "tennis racket": "raquete de tênis",
    "bottle": "garrafa",
    "wine glass": "taça de vinho",
    "cup": "copo",
    "fork": "garfo",
    "knife": "faca",
    "spoon": "colher",
    "bowl": "tigela",
    "banana": "banana",
    "apple": "maçã",
    "sandwich": "sanduíche",
    "orange": "laranja",
    "broccoli": "brócolis",
    "carrot": "cenoura",
    "hot dog": "cachorro-quente",
    "pizza": "pizza",
    "donut": "rosquinha",
    "cake": "bolo",
    "chair": "cadeira",
    "sofa": "sofá",
    "pottedplant": "planta em vaso",
    "bed": "cama",
    "diningtable": "mesa de jantar",
    "toilet": "banheiro",
    "tvmonitor": "televisão",
    "laptop": "laptop",
    "mouse": "mouse",
    "remote": "controle remoto",
    "keyboard": "teclado",
    "cell phone": "celular",
    "microwave": "micro-ondas",
    "oven": "forno",
    "toaster": "torradeira",
    "sink": "pia",
    "refrigerator": "geladeira",
    "book": "livro",
    "clock": "relógio",
    "vase": "vaso",
    "scissors": "tesoura",
    "teddy bear": "urso de pelúcia",
    "hair drier": "secador de cabelo",
    "toothbrush": "escova de dentes"
}

# Função para traduzir o nome da classe
def translate_class_name(class_name):
    return translation_dict.get(class_name, class_name)

# Carregar o modelo pré-treinado ResNet
model = models.resnet50(pretrained=True)
model.eval()

# Transformação para a imagem
preprocess = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize(256),
    transforms.CenterCrop(224),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

# Capturar vídeo da câmera
cap = cv2.VideoCapture(0)

# Verificar se a câmera foi aberta corretamente
if not cap.isOpened():
    print("Erro ao abrir a câmera")
    exit()

# Reduzir a resolução do frame processado para economizar memória
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 640)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 480)

# Processar apenas um a cada N frames
frame_skip = 5
frame_count = 0

# Variável para armazenar o frame processado
processed_frame = None

# Lista para armazenar dados de detecção
detection_data = []

# Função para processar frames em um thread separado
def process_frame():
    global processed_frame, frame_count
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        frame_count += 1
        if frame_count % frame_skip != 0:
            processed_frame = frame
            continue
        
        # Obter as dimensões do frame
        height, width = frame.shape[:2]
        
        # Pré-processar a imagem
        input_tensor = preprocess(frame)
        input_batch = input_tensor.unsqueeze(0)
        
        # Verificar se a GPU está disponível e mover o tensor para a GPU
        if torch.cuda.is_available():
            input_batch = input_batch.to('cuda')
            model.to('cuda')
        
        # Realizar a classificação
        with torch.no_grad():
            output = model(input_batch)
        
        # A classe com a maior pontuação é a classe prevista
        _, predicted_idx = torch.max(output, 1)
        
        # Verificar se o índice previsto está dentro do intervalo da lista de classes
        if predicted_idx.item() < len(classes):
            class_name = classes[predicted_idx.item()]
            translated_class_name = translate_class_name(class_name)
            
            # Armazenar dados de detecção
            current_time = time.time()
            detection_data.append({
                "class": translated_class_name,
                "confidence": torch.nn.functional.softmax(output, dim=1)[0][predicted_idx].item(),
                "time": current_time
            })
            
            # Desenhar o rótulo
            label = f"{translated_class_name}: {torch.nn.functional.softmax(output, dim=1)[0][predicted_idx].item():.2f}"
            cv2.putText(frame, label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
        
        processed_frame = frame

# Iniciar o thread para processar frames
thread = threading.Thread(target=process_frame)
thread.start()

while True:
    if processed_frame is not None:
        # Mostrar o frame processado
        cv2.imshow('Video', processed_frame)
    
    # Sair do loop ao pressionar a tecla 'q'
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Liberar a captura e fechar as janelas
cap.release()
cv2.destroyAllWindows()

# Analisar e visualizar os dados de detecção
df = pd.DataFrame(detection_data)
print(df.head())

# Plotar a distribuição das classes detectadas
plt.figure(figsize=(10, 6))
df['class'].value_counts().plot(kind='bar')
plt.title('Distribuição das Classes Detectadas')
plt.xlabel('Classe')
plt.ylabel('Contagem')
plt.show()
```
